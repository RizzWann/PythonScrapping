```python
from bs4 import BeautifulSoup
import requests
```

```python
import re
```

```python
courses =[]
with open('E:/DS/home.html','r') as home:
    content = home.read()
    soup = BeautifulSoup(content,'lxml')
    cards = soup.find_all('div',class_='card')
    
    for card in cards:
        c_name = card.h5.text
        #c_cost = card.a.text.split()[-1]
        c_cost = re.findall('[0-9]',card.a.text)
        c_cost = ''.join(c_cost)
        #print(f"course:{c_name}, fee:{c_cost}")
        courses.append({
            'course':c_name,
            'fee':int(c_cost)
        })
courses   
```

```python
import pandas as pd
pd.DataFrame(courses)
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
import re
```

```python
html = requests.get('https://www.timesjobs.com/candidate/job-search.html?from=submit&searchType=Home_Search&luceneResultSize=2000&postWeek=60&cboPresFuncArea=35&pDate=Y&sequence=1&startPage=1').text
soup = BeautifulSoup(html,'lxml')
```

```python
jobList = []

jobs = soup.find_all('li',class_='clearfix job-bx wht-shd-bx')
for job in jobs:
    title = re.findall('[a-zA-Z,]',job.h2.a.text)
    title = "".join(title)
    location = job.find('ul',class_='top-jd-dtl clearfix').find('span').text.replace(' ','')
    skills = re.findall('[a-zA-Z,]',job.find('ul',class_='list-job-dtl clearfix').span.text)
    skills = ''.join(skills)
    jobList.append({
        'Role':title,
        'Location':location,
        'Skills':skills 
    })
jobList    
```
